```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../input/nba-player-stats/2021-2022 NBA Player Stats.csv',encoding="Windows-1252", delimiter=";")
dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.columns
```

```python
dataset.drop('Rk',axis=1,inplace=True)
```

Let's find some interseting correlation within the fertures

```python
plt.figure(figsize=(15,15))
sns.heatmap(dataset.corr(),annot=True)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot('Age','PTS',hue='Pos',data=dataset)
plt.xlabel('Age')
plt.ylabel('Points per Game')
plt.title('Age vs Points per game')
```

Usually there is a common trend that points per game will lower as the age increases. but there's an outlier where a player has an age over 35 and point per game close to 30. let's find out who is he. 

```python
dataset[(dataset['Age']>35) & (dataset['PTS'] > 25)]['Player']
```

```python
dataset[(dataset['Age']>25) & (dataset['PTS'] == 29.4)]['Player']
```

It was Lebron James, A Goat for a reason. 

Here is the more refined version of this scatter flow.

```python
plt.figure(figsize=(15,10))
sns.set_style("darkgrid")
sns.scatterplot('Age','PTS',hue='Pos',data=dataset)
plt.xlabel('Age')
plt.ylabel('Points per Game')
plt.title('Age vs Points per Game')
plt.text(dataset.Age[dataset.Player=="LeBron James"]+0.3,dataset.PTS[dataset.Player=="LeBron James"]+0.3,"Lebron James", color='black',bbox=dict(facecolor='yellow',alpha=0.5))
plt.text(dataset.Age[dataset.Player=="Giannis Antetokounmpo"]+0.3,dataset.PTS[dataset.Player=="Giannis Antetokounmpo"]+0.3,"Giannis Antetokounmpo", color='black',bbox=dict(facecolor='yellow',alpha=0.5))
plt.text(dataset.Age[dataset.Player=="Kevin Durant"]+0.3,dataset.PTS[dataset.Player=="Kevin Durant"]+0.3,"Kevin Durant", color='black',bbox=dict(facecolor='yellow',alpha=0.5))
plt.text(dataset.Age[dataset.Player=="Stephen Curry"]+0.3,dataset.PTS[dataset.Player=="Stephen Curry"]+0.3,"Stephen Curry", color='black',bbox=dict(facecolor='yellow',alpha=0.5))
plt.text(dataset.Age[dataset.Player=="James Harden"]+0.3,dataset.PTS[dataset.Player=="James Harden"]+0.3,"James Harden", color='black',bbox=dict(facecolor='yellow',alpha=0.5))
plt.text(dataset.Age[dataset.Player=="Joel Embiid"]-1.8,dataset.PTS[dataset.Player=="Joel Embiid"]-0.8,"Joel Embiid", color='black',bbox=dict(facecolor='yellow',alpha=0.5))
plt.text(dataset.Age[dataset.Player=="Anthony Davis"]+0.3,dataset.PTS[dataset.Player=="Anthony Davis"]+0.3,"Anthony Davis", color='black',bbox=dict(facecolor='yellow',alpha=0.5))

plt.legend(title='Position')

plt.savefig('nba.png')
```
